```python
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib import font_manager
from matplotlib.ticker import MultipleLocator

font_list = font_manager.findSystemFonts(fontpaths=None, fontext="ttf")

font = {"family": "Noto Sans", "weight": "normal", "size": 12}
plt.rc("font", **font)

ORANGE = "#E95420"
LIGH_AUB = "#77216F"
MID_AUB = "#5E2750"
DARK_AUB = "#2C001E"
WARM_AUB = "#772953"
WARM_GREY = "#7a7b7c"
COOL_GREY = "#333333"
DARK_GREY = "#111111"
```

```python
label = [
    "Daily", 
    "Every few days", 
    "Every few weeks", 
    "Every few months",
    "Never", 
]
value = [
    0.02, 
    0.08, 
    0.19, 
    0.42,
    0.29, 
]

df = pd.DataFrame({
    "label": label,
    "value": value,
})
```

```python
fig, ax = plt.subplots(figsize=(8, 6), dpi=300)

plt.barh(y=df["label"], width=df["value"], color=[WARM_GREY] * 2 + [MID_AUB] * 3)
plt.gca().invert_yaxis()

ax.spines[["top", "left", "right", "bottom"]].set_visible(False)
ax.tick_params(axis="y", which="both", left=False, labelsize=14, labelcolor=WARM_GREY)
ax.tick_params(axis="x", which="both", bottom=False)
ax.tick_params(labelbottom=False)

for index, value in enumerate(df["value"]):
    x = value - 0.04 if value > 0.03 else value + 0.004
    y = index
    color = "white" if value > 0.03 else WARM_GREY
    ax.text(x, y, f"{value: .0%}", va="center", ha="left", fontsize=14, color=color)

text = "Drop share-button refactor project"
ax.text(x=-0.125, y=-2.5, s=text, fontsize=32, color=WARM_GREY, fontweight="bold")

text = "no users are using it (or at least very few)"
ax.text(x=-0.125, y=-1.98, s=text, fontsize=20, color=MID_AUB, fontweight="regular")

text = "How often do you share articles from our website?"
ax.text(x=-0.125, y=-1.0, s=text, fontsize=20, color=WARM_GREY, fontweight="bold")

text = f"Most users rarely share articles"
ax.text(x=0.21, y=1.9, s=text, fontsize=14, color=MID_AUB, fontweight="bold")

text = f"with the share button"
ax.text(x=0.21, y=2.2, s=text, fontsize=14, color=MID_AUB, fontweight="regular")

plt.axhline(y=-1.6, xmin=-10, xmax=0.7, color=WARM_GREY, alpha=0.25)

plt.show()
```
